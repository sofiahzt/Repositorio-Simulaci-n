#### 1. Simulación del movimiento browniano

El movimiento browniano es el movimiento aleatorio que se observa en las partículas que se hallan en un medio fluido, como resultado de choques contra las moléculas de dicho fluido [(referencia)].

[(referencia)]: https://es.wikipedia.org/wiki/Movimiento_browniano

Para simular el movimiento, en lugar de usar las ecuaciones integro-diferenciales que describe Einstein, usa una caminta aleatoria para los tres ejes $x$, $y$, $z$ de la siguiente forma
$$
W_i=W_{i-1}+X_i\sqrt{\Delta t}
$$
donde, $\Delta t$ es un intervalo de tiempo y 
$$
X_i=\left\{\begin{array}{l}
1\\
-1
\end{array}\right.,
$$
se escoge de forma aleatoria con la misma probabilidad.

Simule $10$ segundos con un $\Delta t=0.001$ y represéntelo en un gráfico 3D (Puede ver el archivo de Ejemplo 3D).

```python
import random 
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 100

def caminata(T0, N) :
    movimiento = [T0]
    seq = [-1, 1]
    for i in range (N) :
        movimiento.append(movimiento[-1] + random.choice(seq)*np.sqrt(0.001))
    return movimiento

plt.figure()
plt.plot(caminata(0, N))
plt.grid()
```

```python
n = 800

plt.figure()

for i in range(n) :
    plt.plot(caminata(0, N))
    
plt.grid()
```

donde la apuesta es

C_i/2,\quad \mbox{si }C_i\geq 6\\
2,\quad \mbox{de lo contrario.}
\end{array}\right.

```python
def volado() : 
    return np.random.randint(1, 3)
volado()
```

```python
def juego() :
    v = volado()
    gana = False
    if(v == 1) :
        gana = True
    return gana
```

```python
def Apostador(C0, apuesta, N) :
    C = [C0]
    for i in range(N) :
        if juego() :
            C.append(C[-1] + apuesta)
        else :
            C.append(C[-1] - apuesta)
    return C
```

```python
n = 100
plt.figure()
if C >= 6 :
    apuesta = C/2
else:
    apuesta = 2
for i in range(n) :
    plt.plot(Apostador(20, apuesta, 10000))
plt.grid()
```
